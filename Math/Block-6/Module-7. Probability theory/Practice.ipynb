**Задание 4.3**  
Вы едете на деловую встречу в бизнес-центр. Когда на дорогах нет пробок, вероятность не опоздать к назначенному времени составляет для вас 0.9, а когда на улицах пробки, то вероятность успеть вовремя уменьшается до 0.7. Известно, что в том районе, где находится бизнес-центр, пробки бывают в 60 % случаев.

С какой вероятностью вы успеете приехать на деловую встречу без опозданий?

* $ P(TJ) = 0.6 $ - вероятность **наличия пробки** 
* $ P(nTJ) = 0.4 $ - вероятность **отсутствия пробки**
* $ P(nL|nTJ) = 0.9 $ - вероятность **не опоздать**, когда **нет пробок**
* $ P(L|nTJ) = 0.1 $ - вероятность **опоздать**, когда **нет пробок**
* $ P(nL|TJ) = 0.7 $ - вероятность **не опоздать**, когда **есть пробки**
* $ P(L|TJ) = 0.3 $ - вероятность **опоздать**, когда **есть пробки**

$$ P(nL) = P(nL|nTJ) \cdot P(nTJ) + P(nL|TJ) \cdot P(TJ) = 0.9 \cdot 0.4 + 0.7 \cdot 0.6 = 0.78 $$

**Задание 4.4**  
В одном из отделов банка работает 80 % сотрудников с опытом и 20 % новичков. Вероятность, что опытный сотрудник совершит ошибку при проведении банковской операции, равна 0.01. Вероятность, что такую же ошибку совершит новичок, в 50 раз больше.

Найдите вероятность того, что случайная банковская операция будет произведена без ошибок.

* $ P(E) = 0.8 $ - сотрудники **с опытом**
* $ P(nE) = 0.2 $ - сотрудники **без опыта**
* $ P(F|E) = 0.01 $ - вероятность, что **опытный** сотрудник **совершит ошибку**
* $ P(nF|E) = 0.99 $ - вероятность, что **опытный** сотрудник **не совершит ошибку**
* $ P(F|nE) = 0.5 $ - вероятность, что **неопытный** сотрудник **совершит ошибку**
* $ P(nF|nE) = 0.5 $ - вероятность, что **неопытный** сотрудник **не совершит ошибку**

$$ P(nF) = P(nF|E) \cdot P(E) + P(nF|nE) \cdot P(nE) = 0.99 \cdot 0.8 + 0.5 \cdot 0.2 = 0.892 $$

**Задание 4.5**  
Вы выяснили, что 2 % клиентов испытали wow-эффект от вашего сервиса и рассказывают о вас буквально всем — из них 70 % приведут новых клиентов. Ещё 20 % клиентов периодически оставляют вам положительные отзывы в соцсетях и на сайте — из них 10 % приведут новых клиентов. Что думают оставшиеся 78 % клиентов, вы не знаете — из них 4 % приведут новых клиентов.
* $ P(W) = 0.02 $ - клиенты, **испытавшие wow-эффект**
* $ P(N) = 0.2 $ - клиенты, **оставившие положительные отзывы**
* $ P(O) = 0.78 $ - клиенты, о которых **нет информации**
* $ P(C|W) = 0.7 $ - клиенты, которые **испытали wow-эффект** и **привели новых клиентов**
* $ P(C|N) = 0.1 $ - клиенты, которые **оставили положительные отзывы** и **привели новых клиентов**
* $ P(C|O) = 0.04 $ - клиенты, о которых **нет информации** и **привели новых клиентов**

$ P(C) = P(C|W) \cdot P(W) + P(C|N) \cdot P(N) + P(C|O) \cdot P(O) = 0.7 \cdot 0.02 + 0.1 \cdot 0.2 + 0.04 \cdot 0.78 = 0.0652 $ - вероятность, что **случайный** клиент сервиса приведёт **нового клиента**.

$ P(O|C) = \frac{P(C \cap O)}{P(C)} = \frac{P(C|O) \times P(O)}{P(C)} = \frac{0.04 \times 0.78}{0.0652} \approx 0.48 $ - вероятность, что клиент, о котором **нет информации**, привёл **нового клиента**
1. Вычислите апостериорную вероятность принадлежности клиента к классу «wow-эффект», если известно, что он привёл нового клиента.

    $ P(W|C) = \frac{P(C|W) \times P(W)}{P(C)} = \frac{0.7 \times 0.02}{0.0652} \approx 0.21 $
2. Вычислите апостериорную вероятность принадлежности клиента к классу «тихоня», если известно, что он привёл нового клиента.

    $ P(O|C) = \frac{P(C|O) \times P(O)}{P(C)} = \frac{0.04 \times 0.78}{0.0652} \approx 0.48 $
3. Вычислите апостериорную вероятность принадлежности клиента к классу «тихоня», если известно, что он **не** привёл нового клиента.

    $ P(\overline{C}) = 1 - P(C) = 0.9348 $  
    $ P(\overline{C}|O) = 1 - P(C|O) = 0.96 $  
    $ P(O|\overline{C}) = \frac{P(\overline{C}|O) \times P(O)}{P(\overline{C})} = \frac{0.96 \times 0.78}{0.9348} \approx 0.8 $

## <center>Непрерывное распределение

```python
import numpy as np
from scipy.stats import randint
```

Попробуем рассчитать вероятности для случайной величины, которая принимает значения от 1 включительно до 7 не включительно (собственно, это как раз пример с пирожными):

```python
# Задаём возможные значения случайной величины
x = np.arange(1, 7)
# Задаём интервал, на котором будут распределяться наши вероятности
disc_uni_dist = randint(1, 7)
# Вычисляем вероятности выпадения каждого значения случайной величины
pmf = disc_uni_dist.pmf(x)
```

Получаем вероятности для каждого из шести значений:



```python
pmf
```

Также мы можем найти кумулятивную вероятность для распределения, то есть для каждого $x_i$ вероятность того, что случайная величина примет значение $x_i$ или меньше:

```python
# Рассчитаем кумулятивную вероятность
cdf = disc_uni_dist.cdf(x)
print(cdf)
```

**Пример**  
Вы участвуете в розыгрыше бесплатного места на новый курс по Data Science. Вы знаете, что участников 250.

Какова вероятность, что на курс попадёт кто-то из первых пятидесяти зарегистрировавшихся?

```python
# Задаём возможные значения случайной величины
x = np.arange(1, 251)
# Задаём интервал, на котором будут распределяться наши вероятности
disc_uni_dist = randint(1,251)
# Вычисляем вероятности выпадения каждого значения случайной величины
cdf = disc_uni_dist.cdf(x)
# Ищем вероятность того, что на курс попадёт кто-то из первых 50 зарегистрировавшихся
print(cdf[49])
```

В результате получаем $0.2$ — это вероятность того, что выигрыш получит кто-то, чей порядковый номер — от 1 до 50. Разумеется, вероятность не зависит от самого номера, поэтому для номеров с 51 по 100 вероятность будет ровно такая же.

## <center>Распределение Бернулли
С помощью специальной функции `bernoulli.rvs()` из библиотеки `scipy` мы можем смоделировать распределение Бернулли и, например, увидеть ожидаемое соотношение попаданий и промахов для решённой задачи, если будет сделано 500 выстрелов:

```python
from scipy.stats import bernoulli
import seaborn as sns

data = bernoulli.rvs(size=500, p=0.6)
ax = sns.displot(
    data,
    color='dodgerblue',
)
ax.set(xlabel='Значение случайной величины', ylabel='Частота');
```

Также можно вывести количество неудач и попаданий в нашей модели:


```python
unique, counts = np.unique(data, return_counts=True)
print(np.asarray((unique, counts)).T)
```

## <center>Биномиальное распределение
**Пример**  
Производитель гаджетов знает, что 20 % производимых им товаров — бракованные.

Если он производит десять изделий в день, какова вероятность того, что не более двух из них бракованные?


```python
import scipy

scipy.stats.binom.pmf(8,10,0.8) + scipy.stats.binom.pmf(9,10,0.8) + scipy.stats.binom.pmf(10,10,0.8)
```

Как и для распределения Бернулли, для биномиального распределения можно смоделировать какое-то количество попыток и получить результат. Например, можно смоделировать биноминальное распределение с параметром p = 0.5 и количеством испытаний, равным 10, и реализовать 1000 попыток:



```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.binomial(n=10, p=0.5, size=1000)

sns.displot(x)

plt.show()
```

Можно видеть, что чаще всего можно наблюдать пять успешных исходов.
