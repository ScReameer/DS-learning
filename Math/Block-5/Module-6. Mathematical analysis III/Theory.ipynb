# <center>Градиентный спуск: применение и модификации
**Три группы алгоритмов:**
* **методы нулевого порядка / оракул нулевого порядка** (их работа основана на оценке значений самой целевой функции в разных точках);
* **методы первого порядка / оракул первого порядка** (при работе они используют первые производные в дополнение к информации о значении функции);
* **методы второго порядка / оракул второго порядка** (для них необходимо оценивать и значение функции, и значение градиента, и гессиан (матрицу Гессе)).
