# <center>Основные обозначения в математике

1. $\forall$ читается как «для любого» и называется «квантором всеобщности». Он заменяет слова и словосочетания «любой», «всякий», «для любого» и т. п.
2. $\exists$ читается как «существует» и называется «квантором существования».  Он заменяет слова «существует», «найдётся» и т. п.

    $\exists !$ читается как «существует единственный» и называется «квантором существования и единственности».
3. Знак импликации $\implies$ воспринимается как замена слов «следует», «влечёт», «вытекает», «имеет место», «выполняется».
4. Знак равносильности или эквивалентности $\iff$ заменяет словосочетания «тогда и только тогда, когда», «в том и только том случае, когда», «если и только если».
5. Двоеточие $:$ в математических формулах читается как «такой, что».
6. Вертикальная черта $|$ означает «при условии, что».
## Числовые множества
1. $ \mathbb{N} = \{1,2,3,\dots\} $ - все натуральные числа, т. е. все числа, которые мы используем для счёта в обыденной жизни.
2. $ \mathbb{Z} = \{\dots, -2, -1, 0, 1, 2, \dots\} $ - все целые числа: все натуральные числа, противоположные им (т. е. отрицательные) числа и ноль.
3. $ \mathbb{Q} \left\{\frac{p}{q}\right\}, p \in \mathbb{Z}, q \in \mathbb{N} $ - все рациональные числа, т. е. такие, которые можно записать в виде дроби, у которой числитель является целым числом, а знаменатель — натуральным.
4. $ \mathbb{R} $ - все вещественные числа (т. е. не только целые, натуральные и обыкновенные дроби, но также и иррациональные числа, например $\sqrt{2}$).

>**Пример:**  
>$ \exists x \in \mathbb{R} : x : 13 $ расшифровывается как «существуют такие вещественные числа, которые делятся на 13».

## Числовые промежутки
> Числовые промежутки — это числовые множества, которые можно изобразить на координатной прямой.
### ОСНОВНЫЕ ВИДЫ ЧИСЛОВЫХ ПРОМЕЖУТКОВ:
1. **Открытый луч**

    $x<2(-\infty;2)$ - все числа меньше 2.
2. **Закрытый луч**

    $x\geq 5 [-5; +\infty]$ - все числа больше или равны -5.
3. **Отрезок**

    $-5 \leq x \leq 2 [-5; 2] $ - все числа больше или равны -5 и одновременно с этим меньше или равны 2.
4. **Интервал**

    $(-5; 2)$ - все числа больше -5 и одновременно с этим меньше 2.
5. **Полуинтервал**

    $(-5; 2]$ - все числа больше -5 и одновременно с этим меньше или равны 2.
