# <center>Системы линейных уравнений
>**Система уравнений** — это условие, состоящее в одновременном выполнении нескольких уравнений, входящих в систему.
$$
\begin{cases}
    p(x;y)=0,\\
    q(x;y)=0.
\end{cases}
$$
>**Решением** этой системы называется пара чисел $(x;y)$, которая одновременно является решением и первого, и второго уравнения, т. е. при подстановке обращает их оба в верные равенства.

>Если все уравнения, входящие в систему, являются линейными, то представляют собой уравнения вида $ax+by+c=0$, то система называется **линейной**. Если хотя бы одно из уравнений системы не является линейным, то система называется **нелинейной**.


## Метод подстановки
Суть метода заключается в том, что в одном уравнении нужно выразить одну переменную через другую и подставить её во второе уравнение. Таким образом получится одно уравнение с одной неизвестной.
$$
\begin{cases}
    x=5y\\
    2x-7y=6\\
\end{cases}\\
2 \cdot 5y - 7y = 6\\
y=2\\
x=5y=5\cdot2=10\\
(10;2)
$$
## Метод сложения
Суть метода заключается в том, что мы так складываем уравнения системы (левую часть одного уравнения — с левой частью другого, правую часть одного — с правой частью другого), что одна из переменных сокращается. Иногда для этого требуется предварительно умножить одно или оба уравнения на некоторые коэффициенты.
$$
\begin{cases}
    4y+5x=-10\\
    3y-5x=17
\end{cases}\\
4y+5x=-10\\
\frac{3y-5x=17}{7y+0=7}\\
y=1\\
5x=10-4\\
x=-\frac{14}{5}
$$
# <center>Системы линейных неравенств
>Несколько неравенств с одной переменной образуют **систему неравенств**, если требуется найти все общие решения заданных неравенств.

>**Решить систему неравенств** — значит найти те значения переменной, которые являются решениями каждого из неравенств.
$$
\begin{cases}
    2x-1>3\\
    3x-1<11
\end{cases}\\
2x-1 > 3\\
2x>4\\
x>2\\
3x-1 < 11\\
3x<11+1\\
3x<12\\
x<4
$$
