# <center>Неравенства
## Свойства неравенств
1. Если $a>b \spaceи\space b>c, то\space a>c$
2. Если $a>b, то\space a+c>b+c$
3. Если $a>b \spaceи\space m>0, то\space am > bm$
4. Если $a>b \spaceи\space m<0, то\space am < bm$
5. Если $a>b \spaceи\space c>d, то\space a+c>b+d$
6. Если $a,b,c,d>0 \spaceи\space a>b, а\space c>d, то\space ac>bd$
7. Если $a\geq0, b\geq0 \spaceи\space a>b, то\space a^n>b^n, где\space n \in N$
8. Если $a \spaceи\space b$ - числа одного знака и не равны нулю, а также если $a>b, то\space \frac{1}{a} < \frac{1}{b}$
