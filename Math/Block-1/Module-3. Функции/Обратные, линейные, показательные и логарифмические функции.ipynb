## Область определения
>По сути, **область определения** — это множество всех значений аргументов, к которым мы можем применять рассматриваемую функциональную зависимость, и с точки зрения математики это будет «разрешено».

Пусть у нас есть функция $ f(x) = \frac{1}{x} $  
Можно использовать в качестве аргумента любое число, кроме нуля, так как на ноль делить нельзя. Это значит, что областью определения функции являются абсолютно все вещественные числа, за исключением нуля.

Более строго, математически это утверждение можно записать следующим образом:
$$
D(f(x)):x\in(-\infty;0)\cup(0;+\infty)
$$
$D(f(x))$ здесь читается как «область определения функции $f$ от $x$». Утверждение целиком можно прочитать как «Область определения функции $f$ от $x$: $x$  принадлежит интервалу от минус бесконечности до нуля в объединении с интервалом от нуля до плюс бесконечности».  
**Пример:**
$$
f(x) = \frac{\sqrt{x+2}}{x^2-9}
$$
В данном примере необходимо обратить внимание как на числитель, так и на знаменатель. В числителе находится квадратный корень от $x+2$. Корень мы можем извлекать только из неотрицательных чисел, поэтому $x$ должен быть больше или равен $-2$. Выражение в знаменателе принимает нулевое значение при $x=3$ и при $x=-3$.

Таким образом, получаем следующую область определения:
$$
D(f(x)):[-2, 3]\cup(3,+\infty)
$$
Также существуют ограничения для логарифмов: в основании логарифма не может быть неположительного числа или единицы, а показателем логарифма должно быть строго положительное число:
$$
\log_ab:
\begin{cases}
    a>0\\
    a\neq1\\
    b>0
\end{cases}
$$
  
>График не пересекает ось ординат, а лишь бесконечно близко приближается к ней.
## Область значений
Разумеется, не все функции могут принимать какое угодно значение. Например, функция $y=x^2$ может принимать только неотрицательные значения, так как никакое число в квадрате никогда не может быть меньше нуля. Это очень хорошо видно на графике: все точки находятся в верхней (положительной) половине координатной плоскости.  
  
Можно записать это формально следующим образом:
$$
E(f(x)):f(x)\in[0;+\infty]
$$
## Точки пересечения с осями абсции и ординат
Следующий важный шаг в исследовании функции — поиск точек, в которых функция пересекает ось абсцисс (горизонтальная) и ось ординат (вертикальная).

Давайте посмотрим на график ниже. Можно отметить, что в точках, где график пересекает ось $x$, координата по $y$ равна нулю. И наоборот: в точках пересечения с осью ординат координаты по $x$ принимают значение 0.  
  
**Пример:**
$$
f(x) = \frac{x^2-2x}{x-1}\\
\text{Для начала найдём точки пересечения с осью ординат. Для этого просто подставим $0$ вместо $x$:}\\
f(0)=0\\
\text{Таким образом, пересечение с осью ординат находится в точке $(0;0)$}\\
\text{Теперь найдём пересечение с осью $x$. Для этого заменим значение функции на ноль:}\\
0=\frac{x^2-2x}{x-1}\\
x^2-2x=0\\
x_1=0\\
x_2=2\\
\text{Получаем, что точки пересечения с осью абсцисс — $(0;0) и (2;0)$}
$$
## Чётность и нечётность
>Функция является чётной, если для всех значений $x$ выполняется равенство $f(x)=f(-x)$. Это означает, что функция одинакова на положительной и отрицательной полуплоскостях(множество точек плоскости, лежащих по одну сторону от некоторой прямой на этой плоскости) для оси $x$ и симметрична относительно оси $y$.

## Возрастание и убывание
>Функция называется **возрастающей**, если при увеличении её аргумента увеличивается и значение функции.  

  
Формально можно записать это следующим образом:  
$$
x_2 > x_1 \text{ и } f(x_2) > f(x_1) 
$$
>С **убывающей** функцией всё наоборот: её значение уменьшается с увеличением значения аргумента.

  
$$
x_2 > x_1 \text{ и } f(x_2) < f(x_1) 
$$
## Обратная функция
>Обратной называют функцию, которая обращает зависимость, выражаемую рассматриваемой функцией. Например, если функция делает из числа $a$ число $b$, то обратная ей функция делает из числа $b$ число $a$.

Обратные функции бывают полезны, когда известна зависимость первой величины от второй, но нужно узнать, как вторая величина зависит от первой. Например, нам известна площадь для ремонта и, исходя из неё, мы вычисляем необходимые объёмы строительных материалов. Обратная функция позволила бы определить, какую площадь можно отремонтировать с помощью заранее известного объёма строительных материалов.

$$
\text{Пусть у нас есть функция}\\
y=6x-4\\
\text{Выразим из этого равенства $x$}\\
y=6x-4\\
6x=y+4\\
\text{Обратная функция: }x=\frac{y+4}{6} 
$$
## Линейная функция
>Линейной называют функцию, которую можно представить в виде следующей формулы:
>$$ y=kx+b $$
>где $x$ — независимая переменная, $k$ и $b$ — некоторые числа.  



Если число $b$ будет равно 0, то линейная функция будет иметь вид  и одной из её точек всегда будет точка 0 на оси ординат. Эта точка — начало координат. Такую функцию, полученную из функции $y=kx+b$, называют **прямой пропорциональностью**. Графиком прямой пропорциональности является прямая, проходящая через начало координат.


## Обратная пропорциональность
>**Обратной пропорциональностью** называют зависимость, при которой увеличение независимой переменной ведёт к уменьшению значения функции.

В общем виде функция обратной пропорциональности определяется следующим образом:
$$ y=\frac{k}{x} $$
>**Гипербола** — это кривая с двумя ветвями, т. е. кривая, состоящая из двух частей определённой формы.  
  

## Степенная функция
>В своей простейшей форме **степенные функции** описывают ситуации, когда зависимая переменная прямо пропорциональна степени независимой переменной.
>
>Степенная функция имеет следующий вид:
>$$ f(x)=k \cdot x^n $$
>где $k$ и $n$ — вещественные числа.

В зависимости от чётности степени независимой переменной график будет представлять собой:

* гиперболу — для нечётной степени (слева);
* параболу — для чётной степени (справа).

Исключение будет составлять лишь случай, при котором $n=1$: в этой ситуации функция превращается в линейную. Близость к оси $y$ у любого графика зависит от величины коэффициента $n$: чем больше коэффициент, тем быстрее у функции увеличивается значение и тем ближе она к оси $y$.


## Показательная функция
>Показательная функция — это функция вида $f(x)=a \cdot b^x$, где $a$ и $b$ — вещественные числа, причём $b$ — положительное число.

Например:
* $f(x)=2^x$
* $f(x)=(\frac{1}{2})^x$
* $f(x)=3 \cdot e^{2x}$
* $f(x)=5^{4x}$

* Область значений — все положительные числа, так как положительное число в какой-то степени никогда не станет отрицательным числом или нулём: $E(f):(0;+\infty)$
* Не является чётной или нечётной.
## Логарифмическая функция
>Функция, заданная формулой $y=\log_ax,(a>0, a\neq1)$ , называется **логарифмической функцией с основанием** $a$.



Логарифмическая функция обладает следующими свойствами:

* Область определения — все положительные числа, т. е. $D(f):(0;+\infty)$
* Область значений — все вещественные числа , т. е. $E(f):(-\infty;+\infty)$
* Не является чётной или нечётной.
* При основании логарифма $0<a<1$ функция убывает на $D(f):(0;+\infty)$, а при $a>1$ — возрастает на $(0;+\infty)$
