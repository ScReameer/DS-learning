```python
import pandas as pd


orders = pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')

display(orders.head(), products.head())
```

```python
orders_products = orders.merge(
    products,
    how='left',
    left_on='ID товара',
    right_on='Product_ID'
)
orders_products.head()
```

```python
# 8.8
orders_products[orders_products['Name'].isnull()]['Order ID']
```

```python
# 8.9
orders_products[orders_products['Отменен'] == 'Да']['Name'].iloc[0]
```

```python
# 8.10
mask_ispaid = orders_products['Оплачен'] == 'Да'
orders_products['Прибыль'] = orders_products[mask_ispaid]['Price'] * orders_products[mask_ispaid]['Количество']
income_list = orders_products.pivot_table(
    values='Прибыль',
    index='ID Покупателя',
    aggfunc='sum'
)
display(income_list)
display(orders_products.groupby('ID Покупателя')['Прибыль'].sum())
```
